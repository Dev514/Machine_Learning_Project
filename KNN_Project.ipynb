```r
x=data[,-95]
```

```r
y=data[,95]
```

```r
y
```

```r
normalize <- function(x){
  return ((x-min(x))/(max(x)-min(x)))
}
```

```r
x_n=as.data.frame(lapply(x,normalize))
```

```r
random = sample(1:nrow(x),0.8*nrow(x))
```

```r
x_train=x[random,]
x_test=x[-random,]
print(dim(x_train))
print(dim(x_test))
```

```r
y_train=y[random]
y_test=y[-random]
```

```r
install.packages("class")
library(class)
```

```r
prediction <- knn(train = x_train,test = x_test,cl=y_train,k=8)
```

```r
confusion_matrix=table(prediction,y_test)
```

```r
confusion_matrix
```

```r
accuracy <- (sum(diag(confusion_matrix))/sum(rowSums(confusion_matrix))) *100
accuracy
```
